# Feature Processing

```python
y = df['y']
df = df.drop(['y'], axis=1)
df_test = df_test
```

```python
# Remove duplicate columns
from feature_engine.selection import DropDuplicateFeatures
dup_transformer = DropDuplicateFeatures()
dup_transformer.fit(df)
df = dup_transformer.transform(df)
df_test = dup_transformer.transform(df_test)
```

```python
binary_cols = ['x516', 'x556']
```

```python
set(binary_cols).issubset(set(df.columns))
# binary features didn't get removed --> model inputs later?
```

```python
# Remove columns with >15% missing data
missing_perc_train = df.isnull().mean(axis=0)
missing_perc_test = df_test.isnull().mean(axis=0)

cols_to_drop_train = missing_perc_train[missing_perc_train > 0.15].index
cols_to_drop_test = missing_perc_test[missing_perc_test > 0.15].index
df.drop(cols_to_drop_train, axis=1, inplace=True)
df_test.drop(cols_to_drop_test, axis=1, inplace=True)
```

```python
# drop columns <3% variance
from sklearn.feature_selection import VarianceThreshold
selector = VarianceThreshold(0.03)
selector.fit(df)
df = selector.transform(df)
df_test = selector.transform(df_test)
```

```python
# MICE NA Imputation
from fancyimpute import IterativeImputer

imputer = IterativeImputer()

df = imputer.fit_transform(df)
df_test = imputer.transform(df_test)
```
